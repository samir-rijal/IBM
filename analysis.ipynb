```python
region_sales=df.groupby("Region")["Sales_Volume"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,8))
region_sales.plot(kind='bar',color='green',edgecolor='black')
plt.title('Total sales value by region')
plt.xlabel('Region')
plt.ylabel('Total Sales Volume')
plt.xticks(rotation=90)
plt.show()
```

```python
Color_sales=df.groupby("Color")["Sales_Volume"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,8))
region_sales.plot(kind='bar',color='green',edgecolor='black')
plt.title('Total sales value by Color')
plt.xlabel('Color')
plt.ylabel('Total Sales Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_trend = df.groupby("Year")["Price_USD"].mean()

plt.figure(figsize=(10,6))
plt.plot(avg_price_trend.index, avg_price_trend.values,marker='o',linestyle='-',color='blue')
plt.title('Average BMW Price Trend (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Average Price(USD)')
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```

```python
fuel_counts=df["Fuel_Type"].value_counts() 
plt.figure(figsize=(8,8))
plt.pie(fuel_counts,labels=fuel_counts.index,autopct='%1.1f%%')
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
num_cols=["Engine_Size_L","Mileage_KM","Price_USD","Sales_Volume"]  
corr=df[num_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,fmt=".2f",cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
sns.heatmap(df[['Price_USD','Mileage_KM','Engine_Size_L','Sales_Volume']],annot=True,cmap='coolwarm')
plt.title('Heatmap of key numerical features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
total_records =len(df)
avg_price=df["Price_USD"].mean()
total_sales_vol=df["Sales_Volume"].sum()
top_model=df["Model"].value_counts().idxmax()
top_model_sales=df[df["Model"]==top_model]["Sales_Volume"].sum()
best_region=df.groupby("Region")["Sales_Volume"].sum().idxmax()
electric_percentage=(len(df[df["Fuel_Type"]=="Electric"])/total_records)*100
fig,axes=plt.subplots(2,2,figsize=(16,10))
fig.suptitle("BMW Sales Data Analysis Executive Report (2010-2024)",fontsize=20,fontweight="bold",y=0.95)
yearly_sales=df.groupby("Year")["Sales_Volume"].sum()
axes[0,0].plot(yearly_sales.index,yearly_sales.values,marker='s',color='#1C69D2',linewidth=2)
axes[0,0].set_title('Annual Sales Volume Trend',fontsize=14)
axes[0,0].set_ylabel('Total Sales Volume')
axes[0,0].grid(True,alpha=0.7)
region_sales=df.groupby("Region")["Sales_Volume"].sum().sort_values(ascending=False)
axes[0,1].barh(region_sales.index,region_sales.values,color='#28A745')
axes[0,1].set_title('Sales Volume by Region',fontsize=14)
axes[0,1].set_xlabel('Total Sales Volume')
fuel_counts=df["Fuel_Type"].value_counts()
axes[1,0].pie(fuel_counts,labels=fuel_counts.index,autopct='%1.1f%%',colors=['#2c3e50','#2980b9','#7f8c8d','#27ae60'])
axes[1,0].set_title('Fuel Type Distribution',fontsize=14)
sample_df=df.sample(1000)
axes[1,1].scatter(sample_df["Engine_Size_L"],sample_df["Price_USD"],alpha=0.6,color="#643ce7")
axes[1,1].set_title('Price vs Sales Volume (Correlation)',fontsize=14)
axes[1,1].set_xlabel('Price(USD)')
axes[1,1].set_ylabel('Sales Volume')

print("="*60)
print("      BMW Sales Data Analysis Executive Report (2010-2024)")  
print("="*60)
print(f"REPORT SCOPE:                2010-2024") 
print(f"TOTAL RECORDS:             {total_records:,}Units Analyzed")
print("="*60)
print(f"1.FINANCIALS:           Average Price: ${avg_price:,.2f}")
print(f"2.TOP PERFORMER:         The'{top_model}'is the most frequent model in stock.")
print(f"3.MARKET LEADER:            '{best_region}' is the highest performing sales region.")
print(f"4.EV ADOPTION:              Electric vehicles currently make up{electric_percentage:.1f}% of inventory.    ")
print("="*60)
print("CRITICAL INSIGHTS:")
print(f"-The sales volume shows a remarkably stable trend over the  14-year period") 
print(f"-Pricing is evenly distributed,suggesting BMW maintains consistent")
print(f"- presence across both 'Entry-level Luxury'and 'High-end Luxury'tiers.")
print(f"-High overlap in the Scatter Plot indicates that Price does not")
print(f"significantly restrict sales volume in this specific dataset.")
```
